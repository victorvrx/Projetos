```python
def selecionar_pais(nomepais):
    df_pais = df[df.country_or_area.isin(nomepais)].sort_values('year').reset_index()
    df_pais.drop("index",axis=1,inplace=True)
    df_pais.drop("quantity_footnotes",axis=1,inplace=True)
    return df_pais
```

```python
brasil_df = selecionar_pais(['Brazil'])
argentina_df = selecionar_pais(["Argentina"])
uruguay_df = selecionar_pais(["Uruguay"])
Colombia_df = selecionar_pais(["Colombia"])
Peru_df = selecionar_pais(["Peru"])
```

```python
#Hidro eletricity

hidro_brasil_df = selecionar_pais(['Brazil']).query("category=='hydro'")
hidro_argentina_df = selecionar_pais(["Argentina"]).query("category=='hydro'")
hidro_uruguay_df = selecionar_pais(["Uruguay"]).query("category=='hydro'")
hidro_Colombia_df = selecionar_pais(["Colombia"]).query("category=='hydro'")
hidro_Peru_df = selecionar_pais(["Peru"]).query("category=='hydro'")
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))

plt.plot(hidro_brasil_df["year"],hidro_brasil_df["quantity"],label="Brasil", color="green")
plt.plot(hidro_argentina_df["year"],hidro_argentina_df["quantity"], label ="Argentina", color="blue")
plt.plot(hidro_uruguay_df["year"],hidro_uruguay_df["quantity"], label="Uruguay",color = "pink")
plt.plot(hidro_Colombia_df["year"],hidro_Colombia_df["quantity"], label ="Colombia")
plt.plot(hidro_Peru_df["year"],hidro_Peru_df["quantity"], label="Peru")

plt.legend(bbox_to_anchor =(1, 1))
plt.show()
```
